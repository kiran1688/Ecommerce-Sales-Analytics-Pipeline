```sql
--Total Revenue
SELECT 
SUM(product_price) AS total_revenue
FROM ecommerce_catalog.gold.fact_sales;

--Total Orders

SELECT 
COUNT(DISTINCT order_id) AS total_orders
FROM ecommerce_catalog.gold.fact_sales;

--Average Order Value

SELECT 
SUM(payment_value) / COUNT(DISTINCT order_id) AS avg_order_value
FROM ecommerce_catalog.gold.fact_sales;

--Revenue Trend (Monthly)
-- line chart
SELECT 
d.year,
d.month,
SUM(f.product_price + f.freight_value) AS revenue
FROM ecommerce_catalog.gold.fact_sales f
JOIN ecommerce_catalog.gold.dim_date d
ON f.order_date = d.date
GROUP BY d.year, d.month
ORDER BY d.year, d.month;

--Top Selling Products
-- Bar chart
SELECT 
p.product_category_name,
SUM(f.product_price + f.freight_value) AS revenue
FROM ecommerce_catalog.gold.fact_sales f
JOIN ecommerce_catalog.gold.dim_products p
ON f.product_id = p.product_id
GROUP BY p.product_category_name
ORDER BY revenue DESC
LIMIT 10;

--Bar chart
--Sales by Region
SELECT
c.customer_state,
COUNT(DISTINCT f.order_id) AS total_orders
FROM ecommerce_catalog.gold.fact_sales f
JOIN ecommerce_catalog.gold.dim_customers c
ON f.customer_id = c.customer_id
GROUP BY c.customer_state
ORDER BY total_orders DESC;
```
