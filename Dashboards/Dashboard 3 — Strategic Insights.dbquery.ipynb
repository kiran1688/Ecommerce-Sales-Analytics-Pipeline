```sql
--bar chart
--Seller Performance
SELECT 
seller_id,
SUM(product_price) AS revenue,
COUNT(DISTINCT order_id) AS orders
FROM ecommerce_catalog.gold.fact_sales
GROUP BY seller_id
ORDER BY revenue DESC
LIMIT 10;

--counter
--Delivery Performance
SELECT 
AVG(DATEDIFF(order_delivered_customer_date, order_purchase_timestamp)) 
AS avg_delivery_days
FROM ecommerce_catalog.gold.dim_orders;

--counter
--Shipping Cost Impact
SELECT 
AVG(freight_value) AS avg_shipping_cost
FROM ecommerce_catalog.gold.fact_sales;
```
